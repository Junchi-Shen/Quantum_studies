```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

qc_extra_2 = QuantumCircuit(3)
qc_extra_2.h(0)
qc_extra_2.cx(0, 1)
qc_extra_2.cx(1,2)
qc_extra_2.x(1)
qc_extra_2.measure_all()

sampler = StatevectorSampler()
result = sampler.run([qc_extra_2], shots=1024).result()
print(result[0].data.meas.get_counts())


from qiskit.visualization import circuit_drawer
# circuit_drawer(qc_extra_2)
qc_extra_2.draw("mpl", scale=0.6, style="iqp")
```
