```python
from qldpc import codes

# 1. Create a classical seed code (e.g., a Hamming code)
# The library has built-in classical codes to use as building blocks
classical_code = codes.HammingCode(3)

# 2. Generate a Quantum Hypergraph Product Code
# This automatically performs the Kronecker product math to generate H_X and H_Z
quantum_code = codes.HypergraphProductCode(classical_code, classical_code)

# 3. Analyze the code properties
print("Number of physical qubits (n):", quantum_code.num_qubits)
print("Number of logical qubits (k):", quantum_code.dimension)

# You can even calculate or upper-bound the quantum code distance (d)
distance = quantum_code.get_distance()
print(f"Quantum code distance (d): {distance}")
```
